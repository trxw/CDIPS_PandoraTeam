```python
# Rearranging metrics csv file for SQL grouping by date

import pandas as pd
import numpy as np
import datetime as dt
```

```python
df = pd.DataFrame.from_csv('nbs_billboard100.csv',index_col=None)
df.head()
```

```python
tstamps = df.keys()[4:]
dates = pd.to_datetime(df.keys()[4:])
dates
```

```python
df_rearrange = pd.DataFrame()
for i in range(len(df)):
    #i = 0
    print i+1,'/',len(df)
    row = df.iloc[i]
    df_tmp = pd.DataFrame()
    keys = ['artist','artist_id','service','metric']
    cols = list(row[keys])
    counts = row[4:]
    for ts,d,count in zip(tstamps,dates,counts):
        df_tmp = df_tmp.append(pd.DataFrame([cols+[ts,d,count]]
                                   ,columns=keys+['timestamp','datetime','count']) )
        
    df_rearrange = df_rearrange.append(df_tmp)

    
```

```python
df_rearrange.head()
```

```python
df_rearrange.to_csv('nbs_billboard100_rearranged.csv',index=False)
```
